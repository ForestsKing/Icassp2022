```python
import pandas as pd
import numpy as np
import sweetviz as sv
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('./data/data/train_data.csv')
test = pd.read_csv('./data/data/test_data.csv')
```

```python
report = sv.analyze(train)
report.show_html()
```

```python
for col in train.columns[1:-6]:
    plt.figure()
    sns.kdeplot(train[train['Root1']==1][col], shade = True, color='r', label = '1')
    sns.kdeplot(train[train['Root1']==0][col], shade = True, color='b', label = '0')
    plt.xlabel(col)
    plt.legend()
```
